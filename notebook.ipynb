```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load the dataset
ds_jobs = pd.read_csv("customer_train.csv")

# View the dataset
ds_jobs.head()
```

```python
# Copy of ds_jobs for transforming
ds_jobs_transformed = ds_jobs.copy()

ds_jobs_transformed.info()
```

#

### Columns containing categories with only two factors must be stored as Booleans (bool).

```python
# Finding columns with only two factors
ds_jobs_transformed.nunique()[ds_jobs_transformed.nunique() == 2]
```

```python
print(ds_jobs_transformed["relevant_experience"].value_counts())
print()
print(ds_jobs_transformed["job_change"].value_counts())
```

```python
# Creating dictionaries for mapping
relevant_experience_dict = {"Has relevant experience": True, "No relevant experience": False}
job_change_dict = {1: True, 0: False}

# Mapping the values
ds_jobs_transformed["relevant_experience"] = ds_jobs_transformed["relevant_experience"].map(relevant_experience_dict)
ds_jobs_transformed["job_change"] = ds_jobs_transformed["job_change"].map(job_change_dict)
```

```python
# Checking the changes
print(ds_jobs_transformed["relevant_experience"].value_counts())
print()
print(ds_jobs_transformed["job_change"].value_counts())
print()
print(ds_jobs_transformed.dtypes)
```

### Columns containing integers only must be stored as 32-bit integers (int32).

```python
ds_jobs_transformed.dtypes[ds_jobs_transformed.dtypes == "int64"]
```

```python
ds_jobs_transformed["student_id"] = ds_jobs_transformed["student_id"].astype(np.int32)
ds_jobs_transformed["training_hours"] = ds_jobs_transformed["training_hours"].astype(np.int32)
```

### Columns containing floats must be stored as 16-bit floats (float16).


```python
ds_jobs_transformed.dtypes[ds_jobs_transformed.dtypes == "float64"]
```

```python
ds_jobs_transformed["city_development_index"] = ds_jobs_transformed["city_development_index"].astype(np.float16)
```

```python
# Checking the changes
print(ds_jobs_transformed.dtypes)
```

### Columns containing nominal categorical data must be stored as the category data type.

```python
categorical_cols = list(ds_jobs_transformed.columns[ds_jobs_transformed.dtypes == "object"])
ds_jobs_transformed[categorical_cols].head()
```

```python
nominal_cats = ['city', 'gender', 'major_discipline', 'company_type']
ordinal_cats = list(set(categorical_cols) - set(nominal_cats))
print(nominal_cats)
print(ordinal_cats)

for col in categorical_cols:
    ds_jobs_transformed[col] = ds_jobs_transformed[col].astype("category")
```

### Columns containing ordinal categorical data must be stored as ordered categories, and not mapped to numerical values, with an order that reflects the natural order of the column.

```python
for col in ordinal_cats:
    print(col)
    print(ds_jobs_transformed[col].value_counts())
    print()
```

```python
# Creating lists of orders for ordinal categories
enrolled_university_order = ["no_enrollment", "Part time course", "Full time course"]
education_level_order = ["Primary School", "High School", "Graduate", "Masters", "Phd"]
company_size_order = ["<10", "10-49", "50-99", "100-499", "500-999", "1000-4999", "5000-9999", "10000+"]
experience_order = np.append(np.insert(list(np.arange(1, 21).astype(str)), 0, "<1"), ">20")
last_new_job_order = ["never", "1", "2", "3", "4", ">4"]

orders = [enrolled_university_order, education_level_order, company_size_order, experience_order, last_new_job_order]
```

```python
# Reordering the categories
ds_jobs_transformed["enrolled_university"].cat.reorder_categories(
    new_categories = orders[0], 
    ordered = True, 
    inplace = True
)

ds_jobs_transformed["education_level"].cat.reorder_categories(
    new_categories = orders[1], 
    ordered = True, 
    inplace = True
)

ds_jobs_transformed["company_size"].cat.reorder_categories(
    new_categories = orders[2], 
    ordered = True, 
    inplace = True
)

ds_jobs_transformed["experience"].cat.reorder_categories(
    new_categories = orders[3], 
    ordered = True, 
    inplace = True
)

ds_jobs_transformed["last_new_job"].cat.reorder_categories(
    new_categories = orders[4], 
    ordered = True, 
    inplace = True
)
```

```python
# Checking the memory usage difference
print(ds_jobs.memory_usage().sum()/1024, "kb")
print(ds_jobs_transformed.memory_usage().sum()/1024, "kb")
```

### The DataFrame should be filtered to only contain students with 10 or more years of experience at companies with at least 1000 employees, as their recruiter base is suited to more experienced professionals at enterprise companies.

```python
ds_jobs_transformed = ds_jobs_transformed[(ds_jobs_transformed['experience'] >= "10") & (ds_jobs_transformed["company_size"] >= "1000-4999")]
```
